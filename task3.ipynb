## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/astronaut.png` and show it:

```python
astronaut = plt.imread("data/astronaut.png")
plt.imshow(astronaut)
```

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(astronaut, size):
    result = numpy.zeros(astronaut.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(astronaut.shape):
        neighborhood_sum = 0
        if n <= p[0] < astronaut.shape[0]-n and n <= p[1] < astronaut.shape[1]-n :
            for q in numpy.ndindex(size, size):
                neighborhood_sum = astronaut[p[0] + q[0]-n, p[1] + q[1]-n]+neighborhood_sum
            mean = neighborhood_sum/size**2
            result[p] = mean 
    return result
```

**<a id="#">Task 3.1.3.</a>** Test your implementation from above by using the function `meanfilter` for the previously loaded image and show the result:

```python
astronaut_mean = meanfilter(astronaut, 3)
plt.imshow(astronaut_mean)
```
